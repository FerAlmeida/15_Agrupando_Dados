#### Agrupando Valores

```python
dados.head()
```

###### Para cada um dos clientes podemos somar os valores dos pedidos, valores de produtos por motoristas,etc

```python
# Mostrando o valor total de pedidos de cada cliente

dados.groupby("cliente")["valor_pedido"].sum()
```

```python
# Mostrando o valor total de pedidos por motorista

dados.groupby("motorista")["valor_pedido"].sum()
```

```python
# Qual o valor médio de pedido por Estado destino?

dados.groupby("estado_destino")["valor_pedido"].mean()
```

##### * Agrupando por múltiplas colunas

```python
# Agrupando por cliente e estado destino

dados.groupby(["cliente", "estado_destino",])["valor_pedido"].sum()
```

```python
# Melhorando a apresentação dos dados acima com o comando to_frame()

dados.groupby(["cliente", "estado_destino",])["valor_pedido"].sum().to_frame()
```

```python
# Qual é o total de pedidos por cliente e por estado destino

dados.groupby(["cliente", "estado_destino",])["numero_pedido"].count().to_frame()
```

```python
# Qual é o total de pedidos por motorista e por status?

dados.groupby(["motorista", "status"])["numero_pedido"].count().to_frame()
```
